# **MatplotLib**
- Plot points onto a function
- Like importing matlab onto python

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-5,5,100)
y=x**2
plt.plot(x,y)
```

x^2 since its a numpy array, gives a list of x**2 directly rather than for individual elements

````python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-5,5,100)
y=x**2
plt.plot(x,y)
# Joins all the points
````

````python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-5,5,100)
y=x**2
plt.plot(x,y,'o--',color='red')
#Plotting the points

plt.xlabel("x")
# Naming the X axis

plt.ylabel("y")
# Naming the Y axis

plt.legend()
````

## **There's a fgure and Axes in MatPlotLib**

Figure can contain one or more axes



````python
fig,ax=plt.subplots()
print(type(fig))
print(type(ax))
````

The axes can be aligned like a 2D array inside the figure and can be called using indices

````python
X= [2,4,6,8,10]
Y=[1,4,9,19,39]
fig , ax=plt.subplots()
ax.plot(X,Y)
ax.plot(xlabel='x',ylabel='y',title="Test")
plt.show()
````

````python
x=np.arange(-2,2,0.01)
y=x**2
fig,ax=plt.subplots(2,2,figsize=(10,10))
ax[0][0].set_xlabel("new_x")
ax[0][0].set_ylabel("new_y")
ax[0][0].set_title("Single Axis Plot")

ax[0][1].set_xlabel("x")
ax[0][1].set_ylabel("y^.5")
ax[0][1].set_title("Single Axis Plot")

ax[0][0].plot(x,y)
ax[0][1].plot(x,y**.5)
ax[1][0].plot(-x,y)
ax[1][1].plot(x,-y)
plt.show()
````

````python
x=np.arange(-2,2,0.01)
y=x**2
fig,ax=plt.subplots(figsize=(10,10))
ax.set_xlabel("new_x")
ax.set_ylabel("new_y")
ax.set_title("Single Axis Plot")
ax.plot(x,-y)
plt.show()
````

## **We can also do 3D plots**

````python
x=np.arange(-10,10,0.01)
y=x**2
z=x**3
plt.plot(x,y,z)
plt.show()
# Still showing only 2
````

````python
x=np.arange(-10,10,0.01)
y=x**2
z=x**3

fig=plt.figure(figsize=(10,10))
ax = fig.add_subplot(111,projection='3d')
ax.plot(x,y,z)
````

````python
x=np.arange(-2.0,2.0,0.1)
y=np.arange(-2,2.0,0.1)

X,Y=np.meshgrid(x,y)
# Forms a meshgrid
print(X)
plt.plot(X,Y)

Z=X**2+Y**2
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z)
````
