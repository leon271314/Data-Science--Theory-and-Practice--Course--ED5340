## 1. **Create a class called Matrix with methods to perform various matrix operations using the Numpy library. The class should take two matrices as input attributes and implement the following methods:**

a)      elementwise_addition()

b)      elementwise_subtraction()

c)      elementwise_multiplication()

d)      scalar_multiplication(scalar)

e)      determinant()

f)       matrix_multiplication()

g)      transpose()

h)      inverse()

i)       diagonal(): Print the diagonal elements of the first matrix.

j)       rank()

k)      eigen() : Calculate the eigenvalues and eigenvectors of the first matrix

Note: The eigen() method should also plot the eigenvectors using Matplotlib, which should be imported and used inside the Matrix class.

```python
import numpy as np
import matplotlib.pyplot as plt
class Matrix():
    def __init__(self,array):
        self.mat=np.array(array)
    
    def print_mat(self,name,matrix):
        print(f"{name}:")
        print(matrix)
        print()
    
    def elementwise_addition(self,other):
        if self.mat.shape==other.mat.shape:
            temp=Matrix(np.add(self.mat,other.mat))
            self.print_mat("Elementwise Addition",temp.mat)
            return temp
        else:
            raise ValueError('Enter same dimension matrix')
    
    def elementwise_subtraction(self,other):
        if self.mat.shape==other.mat.shape:
            temp=Matrix(np.add(self.mat,-1*other.mat))
            self.print_mat("Elementwise Subtraction",temp.mat)
            return temp
        else:
            raise ValueError('Enter same dimension matrix')
    
    def elementwise_multiplication(self,other):
        if self.mat.shape==other.mat.shape:
            temp=Matrix(np.multiply(self.mat,other.mat))
            self.print_mat("Elementwise Multiplication",temp.mat)
            return temp
        else:
            raise ValueError('Enter same dimension matrix')
    
    def scalar_multiplication(self,scalar):
        temp=Matrix(scalar*np.array(self.mat))
        self.print_mat("Scalar Multiplication",temp.mat)
        return temp
    
    def determinant(self):
        shape=self.mat.shape
        if len(shape)==2 and shape[1]==shape[0]:
            det=np.linalg.det(self.mat)
            self.print_mat("Determinant",round(det,2))
            return det
        else:
            raise ValueError('Enter nxn matrix')
        
    def matrix_multiplication(self,other):
        try:
            temp=Matrix(np.matmul(self.mat,other.mat))
            self.print_mat("Matrix Multiplication",temp.mat)
            return temp
        except ValueError:
            raise ValueError('Enter the right dimensions matrix')
    
    def transpose(self):
        temp=Matrix(np.transpose(self.mat))
        self.print_mat("Transpose",temp.mat)
        return temp.mat

    def inverse(self):
        if len(self.mat.shape)==2:
            temp=Matrix(np.linalg.inv(self.mat))
            self.print_mat("Inverse",temp.mat)
            return temp.mat
        else:
            raise ValueError('Enter 2D square non singular matrix')

    def diagonal(self):
        if len(self.mat.shape)==2 and self.mat.shape[1]==self.mat.shape[0]:
            temp=Matrix(np.diag(self.mat))
            self.print_mat("Diagonal",temp.mat)
            return temp.mat
        else:
            raise ValueError('Enter 2D square matrix')
    
    def rank(self):
        if len(self.mat.shape)==2:
            temp=np.linalg.matrix_rank(self.mat)
            self.print_mat("Rank",temp)
            return temp
        else:
            raise ValueError('Enter 2D matrix')
        
    def eigen(self):
        mat=np.linalg.eig(self.mat)
        temp=Matrix(mat[1])
        print("Eigenvalues",mat[0])
        print("Eigenvectors",temp.mat)
        return mat[0],temp
    def plot_eig(self):
        mat=np.linalg.eig(self.mat)[1]
        #print(mat)
        mat=list(zip(*mat))
        mat=np.array(mat)
        mat2=np.zeros(self.mat.shape)
        plt.quiver(*mat2,mat[:,0],mat[:,1],scale=8)


x=Matrix([[1,2],[3,4]])
y=Matrix([[2,3],[4,4]])

x.elementwise_addition(y)
x.elementwise_subtraction(y)
x.elementwise_multiplication(y)
x.scalar_multiplication(4)
x.determinant()
x.matrix_multiplication(y)
x.transpose()
x.inverse()
x.diagonal()
x.rank()
x.eigen()
x.plot_eig()

```

```python
import matplotlib.pyplot as plt
fig , ax = plt.subplots(3,figsize=(10,10))
x=np.arange(-5,5,0.01)
for i in range(3):
    ax[i].set_xlabel("X")
    ax[i].set_ylabel("Y")

ax[0].plot(x,np.sin(x))
ax[0].set_title("y=Sin(x)")

ax[2].plot(x,np.cos(x))
ax[2].set_title("y=Cos(x)")

ax[1].plot(x,x**2)
ax[1].set_title("y=x^2")

plt.subplots_adjust(left=0.1,bottom=0.1,right=0.9,top=0.9,wspace=0.4,hspace=0.4)
plt.show()
```

## **5. Text File Input Output**

```python
pledge='''India is my country and all Indians are my brothers and sisters.
I love my country and I am proud of its rich and varied heritage.
I shall always strive to be worthy of it.
I shall give respect to my parents, teachers and elders and treat everyone with courtesy.
To my country and my people, I pledge my devotion.
In their well being and prosperity alone, lies my happiness.
Jai Hind!'''

with open("pledge.txt",'w') as file:
    file.write(pledge)

with open("pledge.txt",'r') as file:
    read_pledge=file.read()
    read_pledge=read_pledge.split()
    count={}
    for i in read_pledge:
        if i not in count.keys():
            count[i]=read_pledge.count(i)
    print(count)

import pandas as pd
dataframe_pledge=pd.DataFrame(count.items())
dataframe_pledge.to_csv("data_pledge.txt",index=False)
# Read
pd.set_option('display.max_rows', None)
with open("pledge.txt",'r') as file:
    print(file.read())
    print(pd.read_csv('data_pledge.txt'))
```
